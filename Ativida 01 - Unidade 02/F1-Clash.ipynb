```python
class CarPieces:
    def __init__(self, name, speed, power_unit, cornering, reliabillity, apst):
        self.name = name
        self.speed = speed
        self.power_unit = power_unit
        self.cornering = cornering
        self.reliabillity = reliabillity
        self.apst = apst
        self._team_score = self.speed + self.power_unit + self.cornering +self.reliabillity + (self.apst)/0.02
    def get_team_score(self):
        return self._team_score
    def Description(self):
        print(f"Name: {self.name}\nSpeed: {self.speed}\nPower Unit: {self.power_unit}\nCornering: {self.cornering}\nReliability: {self.reliabillity}\nAVG Pit stop time: {self.apst}")

class Breaks(CarPieces):
    def __init__(self, name, speed, power_unit, cornering, reliabillity, apst):
        super().__init__(name, speed, power_unit, cornering, reliabillity, apst)
    def Description(self):
        super().Description()
class GearBox(CarPieces):
    def __init__(self, name, speed, power_unit, cornering, reliabillity, apst):
        super().__init__(name, speed, power_unit, cornering, reliabillity, apst)
    def Description(self):
        return super().Description()
class RearWing(CarPieces):
    def __init__(self, name, speed, power_unit, cornering, reliabillity, apst):
        super().__init__(name, speed, power_unit, cornering, reliabillity, apst)
    def Description(self):
        return super().Description()
class FrontWing(CarPieces):
    def __init__(self, name, speed, power_unit, cornering, reliabillity, apst):
        super().__init__(name, speed, power_unit, cornering, reliabillity, apst)
    def Description(self):
        return super().Description()
class Suspension(CarPieces):
    def __init__(self, name, speed, power_unit, cornering, reliabillity, apst):
        super().__init__(name, speed, power_unit, cornering, reliabillity, apst)
    def Description(self):
        return super().Description()
class Engine(CarPieces):
    def __init__(self, name, speed, power_unit, cornering, reliabillity, apst):
        super().__init__(name, speed, power_unit, cornering, reliabillity, apst)
    def Description(self):
        return super().Description()

class CarSetup(CarPieces):
    def __init__(self, breaks, gearBox, rearWing, frontWing, suspesion, engine):
        self.breaks = breaks
        self.gearBox = gearBox
        self.rearWing = rearWing
        self.frontWing = frontWing
        self.suspension = suspesion
        self.engine = engine

    def get_team_score(self):
        return (self.breaks.get_team_score() + self.gearBox.get_team_score() +
        self.rearWing.get_team_score() + self.frontWing.get_team_score() +
        self.suspension.get_team_score() + self.engine.get_team_score())
```

```python
def best_bottle(chosen_atributte):
    max_atribute = 0
    list_bottles = []
    for idx, bottle in enumerate(bottles):
        if chosen_atributte in bottles[bottle]:
            if bottles[bottle][chosen_atributte] > max_atribute:
                max_atribute = bottles[bottle][chosen_atributte]

    for idx, bottle in enumerate(bottles):
        if chosen_atributte in bottles[bottle]:
            if bottles[bottle][chosen_atributte] == max_atribute:
                list_bottles.append(list(bottles.keys())[idx])
    return list_bottles

```

```python
from itertools import combinations
# Calcula o somatório das habilidades para um piloto
def calcular_somatorio(piloto):
    return sum(piloto.values())

# Encontra a melhor dupla de pilotos
melhor_somatorio = 0
melhor_dupla = None

# Calcula todas as combinações de pilotos possíveis e encontra a melhor dupla
for piloto1, piloto2 in combinations(pilots.keys(), 2):
    somatorio_habilidades = (calcular_somatorio(pilots[piloto1]) + calcular_somatorio(pilots[piloto2]))//2
    if somatorio_habilidades > melhor_somatorio:
        melhor_somatorio = somatorio_habilidades
        melhor_dupla = (piloto1, piloto2)

print("Melhor dupla de pilotos:", melhor_dupla)
print("Somatório das habilidades da melhor dupla:", melhor_somatorio)
```
